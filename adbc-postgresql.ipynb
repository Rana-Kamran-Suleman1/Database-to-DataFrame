```python
import pandas as pd
import adbc
```

```python
# Database credentials
user = "postgres"
password = "Admin"
host = "localhost"
port = "5432"
database = "MyDatabase"
```

```python
# Create the ADBC connection
conn = adbc.connect(
    "postgresql",
    user=user,
    password=password,
    host=host,
    port=port,
    database=database
)
```

```python
# Load data from the 'customers' table into a pandas DataFrame
df = pd.read_sql("SELECT * FROM customers;", conn)
```

```python
# Display the DataFrame
print(df)

# Close the connection
conn.close()
```
