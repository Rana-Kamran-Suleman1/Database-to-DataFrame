```python
import pandas as pd
from sqlalchemy import create_engine
import psycopg
```

```python
# Database credentials
user = "your_username"       # Replace with your PostgreSQL username
password = "your_password"   # Replace with your PostgreSQL password
host = "localhost"           # Usually localhost
port = "5432"                # Default PostgreSQL port
database = "your_database"   # Replace with the name of your PostgreSQL database
```

```python
# Create the connection to PostgreSQL using psycopg (v3)
conn = psycopg.connect(
    user=user,
    password=password,
    host=host,
    port=port,
    dbname=database
)
```

```python
# Now, create the SQLAlchemy engine using psycopg (v3)
engine = create_engine(f'postgresql+psycopg://{user}:{password}@{host}:{port}/{database}')
```

```python
# Load data from the 'customers' table into a pandas DataFrame
df = pd.read_sql('SELECT * FROM customers;', engine)
```

```python
# Display the DataFrame
print(df)

# Don't forget to close the connection after you're done
conn.close()
```
